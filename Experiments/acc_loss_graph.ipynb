```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
data = pd.read_csv(csv_path)

# Function to annotate each point with its value
def annotate_graph(x, y, ax):
    for i, txt in enumerate(y):
        ax.annotate(f'{txt:.2f}', (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=8)

# Figure for Accuracy with data point values
plt.figure(figsize=(10, 6))
ax1 = plt.gca()  # Get current axis
plt.plot(data['Epoch'], data['Training Accuracy'], label='Train Accuracy', color='blue', marker='o')
plt.plot(data['Epoch'], data['Validation Accuracy'], label='Val Accuracy', color='green', marker='o')
annotate_graph(data['Epoch'], data['Training Accuracy'], ax1)
annotate_graph(data['Epoch'], data['Validation Accuracy'], ax1)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training & Validation Accuracy vs. Epoch')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

# Figure for Loss with data point values
plt.figure(figsize=(10, 6))
ax2 = plt.gca()  # Get current axis
plt.plot(data['Epoch'], data['Training Loss'], label='Train Loss', color='red', marker='o')
plt.plot(data['Epoch'], data['Validation Loss'], label='Val Loss', color='orange', marker='o')
annotate_graph(data['Epoch'], data['Training Loss'], ax2)
annotate_graph(data['Epoch'], data['Validation Loss'], ax2)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training & Validation Loss vs. Epoch')
plt.legend(loc='center right')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
data = pd.read_csv(csv_path)

# Function to annotate each point with its value
def annotate_graph(x, y, ax, offset=(0,10)):
    for i, txt in enumerate(y):
        ax.annotate(f'{txt:.2f}', (x[i], y[i]), textcoords="offset points", xytext=offset, ha='center', fontsize=8)

# Create figure and axis for accuracy
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Training and Validation Accuracy
ax1.plot(data['Epoch'], data['Training Accuracy'], label='Train Accuracy', color='blue', marker='o')
ax1.plot(data['Epoch'], data['Validation Accuracy'], label='Val Accuracy', color='green', marker='o')
annotate_graph(data['Epoch'], data['Training Accuracy'], ax1)
annotate_graph(data['Epoch'], data['Validation Accuracy'], ax1, offset=(0,5))
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_title('Training & Validation Accuracy and Loss vs. Epoch')
ax1.legend(loc=(0.8, 0.57))
ax1.grid(True)

# Create second axis for loss
ax2 = ax1.twinx()

# Plot Training and Validation Loss
ax2.plot(data['Epoch'], data['Training Loss'], label='Train Loss', color='red', marker='o')
ax2.plot(data['Epoch'], data['Validation Loss'], label='Val Loss', color='orange', marker='o')
annotate_graph(data['Epoch'], data['Training Loss'], ax2, offset=(0,-15))
annotate_graph(data['Epoch'], data['Validation Loss'], ax2, offset=(0,-15))
ax2.set_ylabel('Loss', color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.legend(loc='right')

plt.show()
```
