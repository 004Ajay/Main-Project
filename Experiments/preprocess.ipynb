```python
# VIDEO TO FRAMES

import os
import cv2

def split_frames_to_train_val(folder_path, output_folder):
    # Create train and val folders
    train_folder = os.path.join(output_folder, "train")
    train_fight_folder = os.path.join(train_folder, "Fight")
    train_nonfight_folder = os.path.join(train_folder, "No_Fight")

    os.makedirs(train_folder, exist_ok=True)
    os.makedirs(train_fight_folder,exist_ok=True)
    os.makedirs(train_nonfight_folder,exist_ok=True)

    val_folder = os.path.join(output_folder, "val")
    val_fight_folder = os.path.join(val_folder, "Fight")
    val_nonfight_folder = os.path.join(val_folder, "No_Fight")    

    os.makedirs(val_folder, exist_ok=True)
    os.makedirs(val_fight_folder,exist_ok=True)
    os.makedirs(val_nonfight_folder,exist_ok=True)

    # Get all video files in the specified folder
    entries = os.listdir(folder_path)
    for entry in entries:   # train or val
        entry_path = os.path.join(folder_path, entry)

        if os.path.isdir(entry_path):
            sub_folders = os.listdir(entry_path)

            for sub_folder in sub_folders: # fight or no_fight
                sub_path = os.path.join(entry_path, sub_folder)

                if os.path.isdir(sub_path):
                    videos = os.listdir(sub_path)

                    for video in videos:
                        video_path = os.path.join(sub_path, video)
                        video_name = os.path.splitext(video)[0]
                        video1 = cv2.VideoCapture(video_path) # video
                        frame_count = 1
                        success = True
                        while success:
                            success, image = video1.read()
                            if success:
                                if "train" in video_path: # trying to add to 'train' folder by checking if str 'train' present in path
                                    if "No_fight" in video_path:
                                        output_video_folder = os.path.join(train_nonfight_folder,video_name)
                                        os.makedirs(output_video_folder, exist_ok=True) # making directory for storing frames of video
                                        output_path = os.path.join(output_video_folder,
                                                                   f"{video_name}_{frame_count}.jpg")
                                        cv2.imwrite(output_path,image)

                                    elif "Fight" in video_path:
                                        output_video_folder = os.path.join(train_fight_folder, video_name)
                                        os.makedirs(output_video_folder, exist_ok=True)
                                        output_path = os.path.join(output_video_folder,
                                                                   f"{video_name}_{frame_count}.jpg")
                                        cv2.imwrite(output_path,image)

                                elif "val" in video_path:
                                    if "no_fight" in video_path:
                                        output_video_folder = os.path.join(val_nonfight_folder, video_name)
                                        os.makedirs(output_video_folder, exist_ok=True)
                                        output_path = os.path.join(output_video_folder,
                                                                   f"{video_name}_{frame_count}.jpg")
                                        cv2.imwrite(output_path,image)

                                    elif "fight" in video_path:
                                        output_video_folder = os.path.join(val_fight_folder, video_name)
                                        os.makedirs(output_video_folder, exist_ok=True)
                                        output_path = os.path.join(output_video_folder,
                                                                   f"{video_name}_{frame_count}.jpg")
                                        cv2.imwrite(output_path,image)

                                # cv2.imwrite(output_path,image)
                                frame_count = frame_count + 1

                        video1.release()

                        print(f"{video_name} Done\n")

video_folder_path = r"C:\Users\ajayt\OneDrive\Desktop\Main P\HockeyFight"  # path to folder with video
output_folder = r"C:\Users\ajayt\OneDrive\Desktop\Main P\HockeyFrames"  # path to store extracted frame pictures

split_frames_to_train_val(video_folder_path, output_folder)
```

```python
# RESIZE FRAMES to 224x224x3

from PIL import Image
import os

def resize_images(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.lower().endswith(('.jpg', '.jpeg')):
                file_path = os.path.join(root, file)
                try:
                    # Open the image file
                    img = Image.open(file_path)
                    
                    # Resize the image to 224x224
                    resized_img = img.resize((224, 224))
                    
                    # Save the resized image back to the same location
                    resized_img.save(file_path)
                    
                    print(f"Resized and saved: {file_path}")
                except Exception as e:
                    print(f"Error processing {file_path}: {e}")

# Specify the path to your "train" and "val" directories
dir = r"C:\Users\ajayt\OneDrive\Desktop\Main P\HockeyFrames"
train_directory = os.path.join(dir, 'train')
val_directory = os.path.join(dir, 'val')

# Resize images in the "train" directory
resize_images(train_directory)
print("\n\nALL IMAGES IN TRAIN-DIR DONE\n\n")

# Resize images in the "val" directory
resize_images(val_directory)
print("\n\nALL IMAGES IN VAL-DIR DONE\n\n")
```

```python
# SAVING TO NPY: Save resized images in folder with video names in fight and no fight of train
# and val folders to npy files(with folder name as name of folder having frames of the video stored)

from PIL import Image
import numpy as np
import os
import shutil
from tqdm import tqdm

def save_frames_as_npy_and_delete_folders(root_dir, verbose=False):
    # Iterate over 'train' and 'val' directories
    for data_type in ['train', 'val']:
        path = os.path.join(root_dir, data_type)
        if not os.path.exists(path):
            continue

        categories = os.listdir(path)
        for category in tqdm(categories, desc=f'Processing {data_type}'):
            category_path = os.path.join(path, category)
            if not os.path.isdir(category_path):
                continue

            event_folders = os.listdir(category_path)
            for event_folder in tqdm(event_folders, desc=f'Converting {category}', leave=False):
                event_path = os.path.join(category_path, event_folder)
                if not os.path.isdir(event_path):
                    continue

                images = []
                file_names = sorted([f for f in os.listdir(event_path) if f.endswith(('.png', '.jpg'))], key=lambda x: int(x.split('.')[0].split('_')[-1]))
                for file_name in file_names:
                    file_path = os.path.join(event_path, file_name)
                    try:
                        with Image.open(file_path) as img:
                            img_array = np.array(img, dtype=np.float32)  # Changed dtype to float32
                            images.append(img_array)
                    except Exception as e:
                        if verbose:
                            print(f"Error processing {file_path}: {e}")
                        continue

                images_array = np.stack(images, axis=0)
                npy_file_path = os.path.join(category_path, f'{event_folder}.npy')
                np.save(npy_file_path, images_array)
                if verbose:
                    print(f'Saved {npy_file_path} with shape {images_array.shape}')

                try:
                    shutil.rmtree(event_path)
                    if verbose:
                        print(f'Deleted folder {event_path}')
                except Exception as e:
                    if verbose:
                        print(f"Error deleting {event_path}: {e}")

# Use the function
dataset_root = r"C:\Users\ajayt\OneDrive\Desktop\Main P\HockeyFramesFloat"
save_frames_as_npy_and_delete_folders(dataset_root, verbose=True)
```
