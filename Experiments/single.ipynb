```python
import numpy as np
import matplotlib.pyplot as plt

# Load the npy file
data = np.load(test)
data = data / 255.0 # normalizing    

# Display basic information about the loaded data
data_info = {
    'data_type': data.dtype,
    'shape': data.shape,
    'number_of_frames': data.shape[0] if data.ndim >= 3 else 'Not a video'
}

print(data_info)

# Function to plot frames
def plot_frames(data, frame_indices, titles):
    fig, axes = plt.subplots(1, len(frame_indices), figsize=(15, 5))
    for ax, frame_index, title in zip(axes, frame_indices, titles):
        ax.imshow(data[frame_index])
        ax.set_title(f"{title} Frame")
        ax.axis('off')
    plt.show()

# Frame indices to plot: first, middle, and last
total_frames = data.shape[0]
first_frame = 0
middle_frame = total_frames // 2
last_frame = total_frames - 1
frame_indices = [first_frame, middle_frame, last_frame]
titles = ['First', 'Middle', 'Last']

# Plot the frames
plot_frames(data, frame_indices, titles)
```
