```python
import matplotlib.pyplot as plt

# Data for plotting
percent_categories = [
    "Increase in civilian fatalities vs. 2022", "Rise in explosive weapon use",
    "Civilian fatalities from air strikes", "Civilians harmed in towns/cities", 
    "Air-launched attacks", "Ground-launched attacks", "Civilian casualties from state actors",
    "Violence incidents in populated areas", "Civilians harmed in populated areas"
]
percent_values = [122, 69, 67, 90, 226, 56, 77, 76, 90]

# Sorting the data based on values
sorted_indices = sorted(range(len(percent_values)), key=lambda k: percent_values[k])
sorted_percent_categories = [percent_categories[i] for i in sorted_indices]
sorted_percent_values = [percent_values[i] for i in sorted_indices]

# Creating the plot
plt.figure(figsize=(8, 6))
plt.barh(sorted_percent_categories, sorted_percent_values, color='skyblue')
plt.xlabel('Percentage')
plt.title('Sorted Percentages of Various Factors')
plt.tight_layout()
plt.show()
```
