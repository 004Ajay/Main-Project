```python
import os
from collections import defaultdict
import cv2
import matplotlib.pyplot as plt

def count_video_resolutions(root_dir):
    resolution_count = defaultdict(int)

    # Walk through all subdirectories
    for subdir, dirs, files in os.walk(root_dir):
        for file in files:
            if file.endswith(('.mp4', '.avi', '.mkv')):  # Filter video files
                try:
                    file_path = os.path.join(subdir, file)
                    vid = cv2.VideoCapture(file_path)
                    height = vid.get(cv2.CAP_PROP_FRAME_HEIGHT)
                    width = vid.get(cv2.CAP_PROP_FRAME_WIDTH)
                    resolution = f"{int(width)}x{int(height)}"
                    resolution_count[resolution] += 1
                    vid.release()
                except Exception as e:
                    print(f"Error processing file: {file_path}, Error: {e}")

    # Filter resolutions with count less than 10
    resolution_count_filtered = {resolution: count for resolution, count in resolution_count.items() if count >= 10}

    return resolution_count_filtered

def plot_resolution_frequency(resolution_count):
    resolutions = list(resolution_count.keys())
    counts = list(resolution_count.values())
    plt.figure(figsize=(10, 6))
    plt.bar(resolutions, counts, color='skyblue')
    plt.xlabel('Resolution')
    plt.ylabel('Frequency')
    plt.title('RWF-2000 Video Resolution Frequency')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    # Replace 'root_directory' with the path to your dataset directory
    root_directory = 'D:/Git-Uploads/Main Project/Main-Project/violence_data'

    resolutions_count = count_video_resolutions(root_directory)
    plot_resolution_frequency(resolutions_count)
```

```python
import os
from collections import defaultdict
import cv2
import matplotlib.pyplot as plt

def count_video_resolutions(root_dir):
    resolution_count = defaultdict(int)

    # Walk through all subdirectories
    for subdir, dirs, files in os.walk(root_dir):
        for file in files:
            if file.endswith(('.mp4', '.avi', '.mkv')):  # Filter video files
                try:
                    file_path = os.path.join(subdir, file)
                    vid = cv2.VideoCapture(file_path)
                    height = vid.get(cv2.CAP_PROP_FRAME_HEIGHT)
                    width = vid.get(cv2.CAP_PROP_FRAME_WIDTH)
                    resolution = f"{int(width)}x{int(height)}"
                    resolution_count[resolution] += 1
                    vid.release()
                except Exception as e:
                    print(f"Error processing file: {file_path}, Error: {e}")

    # Filter resolutions with count less than 10
    resolution_count_filtered = {resolution: count for resolution, count in resolution_count.items() if count >= 10}

    return resolution_count_filtered

def plot_resolution_frequency(resolution_count):
    sorted_resolutions = sorted(resolution_count.items(), key=lambda x: x[1], reverse=True)
    resolutions = [res[0] for res in sorted_resolutions]
    counts = [res[1] for res in sorted_resolutions]

    plt.figure(figsize=(10, 6))  # Adjust width and height as needed
    bars = plt.bar(resolutions, counts, color='skyblue')

    # Display frequency count on top of each bar
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, str(int(height)), ha='center', va='bottom')

    plt.xlabel('Resolution')
    plt.ylabel('Frequency')
    plt.title('Video Resolution Frequency')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    # Replace 'root_directory' with the path to your dataset directory
    root_directory = 'D:/Git-Uploads/Main Project/Main-Project/violence_data'

    resolutions_count = count_video_resolutions(root_directory)
    plot_resolution_frequency(resolutions_count)
```
